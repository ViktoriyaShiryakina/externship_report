```python
#import packages
import os
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

ini_data = pd.ExcelFile('ПД_Сектор роста_WRK.xlsm')
df = ini_data.parse('Data')
```

```python
#отбираем только оставшиеся после IV 12 фин коэффициентов
findata = df[['Eq_TL', 'Eq_TA', 'Cash_TA', 'Quick_rat', 'ROE', 
             'EBIT_IntEx', 'AccPay_S', 'S_TA', 'CFO_S', 'CFO_OperPr', 'CFO_Eq', 'lnTA']]
findata
```

```python
#находим значение 5 квантиля

quant5_1 = findata['Eq_TL'].quantile(0.05, interpolation = 'lower')
quant5_2 = findata['Eq_TA'].quantile(0.05, interpolation = 'lower')
quant5_3 = findata['Cash_TA'].quantile(0.05, interpolation = 'lower')
quant5_4 = findata['Quick_rat'].quantile(0.05, interpolation = 'lower')
quant5_5 = findata['ROE'].quantile(0.05, interpolation = 'lower')
quant5_6 = findata['EBIT_IntEx'].quantile(0.05, interpolation = 'lower')
quant5_7 = findata['AccPay_S'].quantile(0.05, interpolation = 'lower')
quant5_8 = findata['S_TA'].quantile(0.05, interpolation = 'lower')
quant5_9 = findata['CFO_S'].quantile(0.05, interpolation = 'lower')
quant5_10 = findata['CFO_OperPr'].quantile(0.05, interpolation = 'lower')
quant5_11 = findata['CFO_Eq'].quantile(0.05, interpolation = 'lower')
quant5_12 = findata['lnTA'].quantile(0.05, interpolation = 'lower')
```

```python
#заменяем нижние выбросы в findata на 5 квантиль
import warnings
warnings.filterwarnings('ignore')

findata['Eq_TL'].loc[findata['Eq_TL'] < quant5_1] = quant5_1
findata['Eq_TA'].loc[findata['Eq_TA'] < quant5_2] = quant5_2
findata['Cash_TA'].loc[findata['Cash_TA'] < quant5_3] = quant5_3
findata['Quick_rat'].loc[findata['Quick_rat'] < quant5_4] = quant5_4
findata['ROE'].loc[findata['ROE'] < quant5_5] = quant5_5
findata['EBIT_IntEx'].loc[findata['EBIT_IntEx'] < quant5_6] = quant5_6
findata['AccPay_S'].loc[findata['AccPay_S'] < quant5_7] = quant5_7
findata['S_TA'].loc[findata['S_TA'] < quant5_8] = quant5_8
findata['CFO_S'].loc[findata['CFO_S'] < quant5_9] = quant5_9
findata['CFO_OperPr'].loc[findata['CFO_OperPr'] < quant5_10] = quant5_10
findata['CFO_Eq'].loc[findata['CFO_Eq'] < quant5_11] = quant5_11
findata['lnTA'].loc[findata['lnTA'] < quant5_12] = quant5_12
```

```python
#находим значение 95 квантиля

quant95_1 = findata['Eq_TL'].quantile(0.95, interpolation = 'higher')
quant95_2 = findata['Eq_TA'].quantile(0.95, interpolation = 'higher')
quant95_3 = findata['Cash_TA'].quantile(0.95, interpolation = 'higher')
quant95_4 = findata['Quick_rat'].quantile(0.95, interpolation = 'higher')
quant95_5 = findata['ROE'].quantile(0.95, interpolation = 'higher')
quant95_6 = findata['EBIT_IntEx'].quantile(0.95, interpolation = 'higher')
quant95_7 = findata['AccPay_S'].quantile(0.95, interpolation = 'higher')
quant95_8 = findata['S_TA'].quantile(0.95, interpolation = 'higher')
quant95_9 = findata['CFO_S'].quantile(0.95, interpolation = 'higher')
quant95_10 = findata['CFO_OperPr'].quantile(0.95, interpolation = 'higher')
quant95_11 = findata['CFO_Eq'].quantile(0.95, interpolation = 'higher')
quant95_12 = findata['lnTA'].quantile(0.95, interpolation = 'higher')
```

```python
#заменяем верхние выбросы в findata на 95 квантиль

findata['Eq_TL'].loc[findata['Eq_TL'] > quant95_1] = quant95_1
findata['Eq_TA'].loc[findata['Eq_TA'] > quant95_2] = quant95_2
findata['Cash_TA'].loc[findata['Cash_TA'] > quant95_3] = quant95_3
findata['Quick_rat'].loc[findata['Quick_rat'] > quant95_4] = quant95_4
findata['ROE'].loc[findata['ROE'] > quant95_5] = quant95_5
findata['EBIT_IntEx'].loc[findata['EBIT_IntEx'] > quant95_6] = quant95_6
findata['AccPay_S'].loc[findata['AccPay_S'] > quant95_7] = quant95_7
findata['S_TA'].loc[findata['S_TA'] > quant95_8] = quant95_8
findata['CFO_S'].loc[findata['CFO_S'] > quant95_9] = quant95_9
findata['CFO_OperPr'].loc[findata['CFO_OperPr'] > quant95_10] = quant95_10
findata['CFO_Eq'].loc[findata['CFO_Eq'] > quant95_11] = quant95_11
findata['lnTA'].loc[findata['lnTA'] > quant95_12] = quant95_12
```

```python
#Расчет граничных значений фин коэффициентов

mindata = findata.min()
mindata = mindata.to_frame()
mindata = mindata.rename(columns={0: "X_min"})

maxdata = findata.max()
maxdata = maxdata.to_frame()
maxdata = maxdata.rename(columns={0: "X_max"})

meandata = findata.mean()
meandata = meandata.to_frame()
meandata = meandata.rename(columns={0: "X_mean"})

stddata = findata.std()
stddata = stddata.to_frame()
stddata = stddata.rename(columns={0: "X_std"})

boundaries = pd.concat([mindata, maxdata, meandata, stddata], axis = 1)
boundaries
```

```python
#создаем копию findata для конечного нормирования значений
findata_copy = df[['Eq_TL', 'Eq_TA', 'Cash_TA', 'Quick_rat', 'ROE', 
             'EBIT_IntEx', 'AccPay_S', 'S_TA', 'CFO_S', 'CFO_OperPr', 'CFO_Eq', 'lnTA']]
findata_copy
```
